The aim of this model is to predict the GPA of students from their SAT scores using machine learning(Linear Regression), Below I:

- Imported the necessary libraries
- Split the dataset into the matrix of features and dependent vector
- Split the matrix of features and dependent vectors into training and test sets
- Trained the model
- Used the model to make a description based on the test set
- Used a plot to visualise the model and the actual training dataset
- Used a plot to visualize the model and the test dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv("/home/darasimi/Downloads/101-simple-linear-regressioncsv/1.01. Simple linear regression.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)

"""from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)"""

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

print(y_test)
print('\n')
print(y_pred)
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('SAT Scores vs GPA (Training set)')
plt.xlabel('SAT Scores')
plt.ylabel('GPA')
plt.show()
```

```python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('SAT Scores vs GPA (Test set)')
plt.xlabel('SAT Scores')
plt.ylabel('GPA')
plt.show()
```
